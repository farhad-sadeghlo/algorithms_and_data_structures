#### Starter Code for Question 1

```python
from collections import deque
from typing import Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(
        self,
        val: int = 0,
        left: Optional["TreeNode"] = None,
        right: Optional["TreeNode"] = None,
    ) -> None:
        self.val = val
        self.left:Optional["TreeNode"] = left
        self.right:Optional["TreeNode"] = right
    
def is_duplicate(root:Optional[TreeNode]) -> int:
    if root is None:
        return -1 
    seen: set[int] = set()
    queue = deque([root])
    
    while queue:
        node = queue.popleft()
        
        if node.val in seen:
            return node.val
        seen.add(node.val)
        
        if node.left is not None:
            queue.append(node.left)
        
        if node.right is not None:
            queue.append(node.right)
        
    return -1

# Time complexity: O(n), because each node is visited once.
# Space complexity: O(n), beacuse the queue and seen set can store up to n nodes/values in the worst case.
```

```python
root1 = TreeNode(val=1, 
                left=TreeNode(val=2, left=TreeNode(val=3), right=TreeNode(val=5)),
                right=TreeNode(val=2, left=TreeNode(val=6), right=TreeNode(val=7)))
print(f"First example result is: {is_duplicate(root1)}")

root2 = TreeNode(val=1, 
                left=TreeNode(val=10, left=TreeNode(val=3), right=TreeNode(val=10)),
                right=TreeNode(val=2, left=TreeNode(val=12), right=TreeNode(val=12)))
print(f"Second example result is: {is_duplicate(root2)}")

root3 = TreeNode(val=10,
                 left=TreeNode(val=9, left=TreeNode(val=8)),
                 right=TreeNode(val=7))
print(f"Third example result is: {is_duplicate(root3)}")
```
